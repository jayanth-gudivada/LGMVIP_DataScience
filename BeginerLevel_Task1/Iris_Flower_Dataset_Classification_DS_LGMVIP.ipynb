# Step-6 : Data Preparation for Model building

```python
# Remove rows with missing values
iris.dropna(inplace=True)

# Splitting the dataset into training and testing sets
X = iris.iloc[:, :-1]
y = iris.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Step-7 : Traning Model

**Using LogisticRegression :**

```python
# Training the logistic regression model

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X,y) #coef of prediction
```

```python
model.coef_
```

```python
model.intercept_
```

# Step-8 : Making Predictions

```python
# Predicting the output value for a single test case
X_test_case = pd.DataFrame({'sepal_length_in_cm': [4.3], 'sepal_width_in_cm': [3], 'petal_length_in_cm': [1.1], 'petal_width_in_cm': [0.1]})
y_pred = model.predict(X_test_case)

# Print the predicted output value
print(y_pred[0])
```

# Step-9 : Summary Statistics

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()

# Convert to pandas dataframe
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                  columns=iris['feature_names'] + ['target'])

# Compute summary statistics
summary = df.describe()
corr = df.corr()['target'][:-1]

# Print the summary statistics
print("Summary Statistics:")
print("{:>15} {:>4} {:>6} {:>6} {:>16}".format("", "Min", "Max", "Mean", "SD", "Class Correlation"))
for col in summary.columns:
    if col != 'target':
        print("{:>15} {:>4.1f} {:>4.1f} {:>6.2f} {:>6.2f} {:>16.4f}".format(col, summary.loc['min', col],
                                                                            summary.loc['max', col],
                                                                            summary.loc['mean', col],
                                                                            summary.loc['std', col],
                                                                            corr[col]))
```
